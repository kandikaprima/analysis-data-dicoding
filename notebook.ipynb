## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana perbandingan curah hujan pada kota Aotizhongxin dan kota changping pada tahun 2016?

```python
plt.figure(figsize=(10,5))
plt.plot(monthly_recap_rain_Aotizhongxin["month"], monthly_recap_rain_Aotizhongxin["RAIN"], marker='o', label="Aotizhongxin")
plt.plot(monthly_recap_rain_Changping["month"], monthly_recap_rain_Changping["RAIN"], marker='o', label="Changping")
plt.title('Perbandingan Curah Hujan pada Kota Aotizhongxin dan Changping di tahun 2016')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

### Pertanyaan 2: Bagaimana perbandingan temperatur pada kota Aotizhongxin dan kota changping pada tahun 2016?

```python
plt.figure(figsize=(10,5))
plt.plot(monthly_recap_temp_Aotizhongxin["month"], monthly_recap_temp_Aotizhongxin["TEMP"], marker='o', label="Aotizhongxin")
plt.plot(monthly_recap_temp_Changping["month"], monthly_recap_temp_Changping["TEMP"], marker='o', label="Changping")
plt.title('Perbandingan Temperatur pada Kota Aotizhongxin dan Changping di tahun 2016')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

### Pertanyaan 2: Apakah curah hujan mempengaruhi temperatur pada tahun 2016 di kota Aotizhongxin?

```python
plt.scatter(monthly_recap_rain_temp_Aotizhongxin["RAIN"], monthly_recap_rain_temp_Aotizhongxin["TEMP"])
plt.title('Perbandingan curah hujan dengan temperatur di kota Aotizhongxin')
plt.xlabel('Curah Hujan')
plt.ylabel('Temperatur')
plt.show()
```

## Conclusion

1. Bagaimana perbandingan curah hujan pada kota Aotizhongxin dan kota changping pada tahun 2016?
   - Pada bulan 7 dan 8 terjadi perbedaan curah hujan dengan kota Aotizhongxin memiliki curah hujan yang lebih tinggi.
2. Bagaimana perbandingan temperatur pada kota Aotizhongxin dan kota changping pada tahun 2016?
   - Relatif memiliki temperatur yang sama pada kedua kota, dengan kota Aotizhongxin sedikit lebih tinggi pada bulan 3 hingga bulan 12.
3. Apakah curah hujan mempengaruhi temperatur pada tahun 2016 di kota Aotizhongxin dan kota changping?
   - Curah hujan memperngaruhi temperatur dimana semangkin tinggi curah hujan maka temperatur juga semakin tinggi.
